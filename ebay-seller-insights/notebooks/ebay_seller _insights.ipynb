```python
def clean_columns(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_", regex=False)
        .str.replace("-", "_", regex=False)
        .str.replace("/", "_", regex=False)
        .str.replace("(", "", regex=False)
        .str.replace(")", "", regex=False)
        .str.replace(",", "", regex=False)
    )
    return df

orders = clean_columns(orders)
financials = clean_columns(financials)
orders.head()
```

```python
def clean_currency(series):
    return (
        series.astype(str)
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.strip()
        .replace({"": None, "nan": None})
        .astype(float)
    )
```

```python
orders["sold_for"] = clean_currency(orders["sold_for"])
orders["shipping_and_handling"] = clean_currency(orders["shipping_and_handling"])
orders["ebay_collected_tax"] = clean_currency(orders["ebay_collected_tax"])
orders["total_price"] = clean_currency(orders["total_price"])
money_cols = [
    "total_sales_includes_taxes",
    "item_sales",
    "taxes_and_government_fees_paid_by_buyer_to_you",
    "taxes_and_government_fees_paid_by_buyer_to_ebay",
    "shipping_and_handling_paid_by_buyer_to_you",
    "total_selling_costs",
    "final_value_fees",
    "promoted_listings___general_fees",
    "shipping_labels_cost_amount_you_paid_to_buy_shipping_labels_on_ebay",
    "net_sales_net_of_taxes_and_selling_costs",
    "average_selling_price"
]

for col in money_cols:
    financials[col] = clean_currency(financials[col])
```

```python
# 1. Remove junk/footer rows
orders_clean = orders_clean[
    orders_clean["order_number"].notna() &
    (orders_clean["order_number"] != "record(s) downloaded")
].copy()

# 2. Drop rows missing required order/date fields
orders_clean = orders_clean.dropna(subset=["order_number", "sale_date"]).copy()

# 3. Fill missing shipping with 0 temporarily
orders_clean["shipping_and_handling"] = orders_clean["shipping_and_handling"].fillna(0)

# 4. Create allocated shipping per item within each order
orders_clean["shipping_allocated"] = (
    orders_clean.groupby("order_number")["shipping_and_handling"].transform("sum") /
    orders_clean.groupby("order_number")["order_number"].transform("count")
)

# 5. Replace shipping column with allocated shipping
orders_clean["shipping_and_handling"] = orders_clean["shipping_allocated"]

# 6. Drop helper column
orders_clean = orders_clean.drop(columns=["shipping_allocated"])

# 7. Drop rows with no item title (this removes the order summary row)
orders_clean = orders_clean.dropna(subset=["item_title"]).copy()

# 8. Fill missing buyer state
orders_clean["buyer_state"] = orders_clean["buyer_state"].fillna("Unknown")

# 9. Recalculate gross revenue
orders_clean["gross_revenue"] = (
    orders_clean["sold_for"] + orders_clean["shipping_and_handling"]
)

# 10. Check results
print(orders_clean.isna().sum())
print(orders_clean.shape)
orders_clean.head()
```
